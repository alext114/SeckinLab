```python
## Initialisation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df= pd.read_csv('C:/Users/Morty/Desktop/netflow_day-02.csv', nrows=1000000)
X = df[['SrcPackets','DstPackets','SrcBytes','DstBytes']]
colmap = {1: 'r', 2: 'g', 3: 'b'}
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_
```

```python


colors = list(map(lambda x: colmap[x+1], labels))

plt.scatter(X['SrcPackets'], X['SrcBytes'], color=colors, alpha=0.5, edgecolor='k')
for idx, centroid in enumerate(centroids):
    #Centroid[0]=SrcPackets
    #Centroid[1]=DstPackets
    #Centroid[2]=SrcBytes
    #Centroid[3]=DstBytes
    
    plt.scatter(centroid[0],centroid[2], color=colmap[idx+1])
plt.xlabel('Source Packets')
plt.ylabel('Source Bytes')
plt.title('Source Packets vs Source Bytes')
plt.show()
```
